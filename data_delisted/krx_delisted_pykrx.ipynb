```python
from pykrx import stock
import FinanceDataReader as fdr
```

```python
# KRX stock delisting symbol list 상장폐지 종목 전체 리스트
krx_delisting = fdr.StockListing('KRX-DELISTING')
krx_delisting
```

```python
krx_delisting = krx_delisting[krx_delisting.SecuGroup.isin(['주권'])].reset_index(drop=True)
```

```python
krx_delisting.info()
```

```python
start_date = '2000-01-01'
end_date = '20220331'
```

```python
from tqdm import tqdm
no_data_list = []
for ticker in tqdm(krx_delisting.Symbol):
    try:
        stock.get_market_ohlcv_by_date(fromdate=start_date, 
                                       todate=end_date,
                                       ticker=ticker).to_parquet(f'data_delisted_pykrx/{ticker}.parquet')
    except:
        no_data_list += [ticker]
```

```python
import os
files = [f for f in os.listdir('data_delisted') if len(f)==14] # 종목 데이터만 선택
```

```python
import pandas as pd
pd.read_parquet(f'data_delisted/{files[0]}')
```

```python
data = pd.concat([pd.read_parquet(f'data_delisted/{file}') for file in files], axis=0)
```

```python
data.to_parquet('krx_delisted_pykrx.parquet')
```
