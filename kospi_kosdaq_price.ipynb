# 2. 상장종목의 주가 조회

+ 삼성전자(005930)의 주가 조회
+ [삼성전자의 상장일은 1975-06-11](http://asp01.fnguide.com/SVO2/asp/SVD_Corp.asp?pGB=1&gicode=A005930&cID=&MenuYn=Y&ReportGB=&NewMenuID=102&stkGb=752)
+ 각 API를 통해 어느 시점부터 데이터를 조회할 수 있는지 비교

```python
start_date = '1975-06-11'
end_date = '2022-03-23'
```

## 2.1. FinanceDataReader

+ 시간이 지날수록 조회할 수 있는 최대 과거 시점이 점점 뒤로 밀림
+ 예를 들어, 2022-03-22에 조회했을 때는 1998-02-05 부터 조회됐었는데, 2022-03-23에 조회했을 때는 1998-02-06 부터 조회됨
+ 수정주가로 조회됨

```python
df_fdr = fdr.DataReader('005930', start=start_date, end=end_date)
df_fdr
```

## 2.2. pykrx

+ 1990년 데이터부터 조회 가능
+ 수정주가가 디폴트로 조회됨(`adjsted`=True)
+ `adjusted`=False로 조회하면 1995-05-02부터 조회되고, 거래대금과 등락률 칼럼도 붙어있음

```python
df_pykrx = stock.get_market_ohlcv_by_date(fromdate=start_date, 
                                          todate=end_date,
                                          ticker="005930")
df_pykrx
```

```python
df_pykrx = stock.get_market_ohlcv_by_date(fromdate=start_date, 
                                          todate=end_date,
                                          ticker="005930", adjusted=True)
df_pykrx
```

```python
df_pykrx = stock.get_market_ohlcv_by_date(fromdate=start_date, 
                                          todate=end_date,
                                          ticker="005930", adjusted=False)
df_pykrx
```

## 2.3. pandas_datareader

+ 데이터소스를 naver로 지정하면 1990년 데이터부터, yahoo로 지정하면 2000년 데이터부터 조회됨
+ 수정주가로 조회되는데 adjust 옵션은 따로 없는 듯
+ yahoo finance를 데이터 소스로 사용할 때는 종목코드 뒤에 코스피 종목인 경우 .KS, 코스닥 종목인 경우 .KQ를 붙여줘야 함

```python
df_pdr = pdr.DataReader('005930', 'naver', start=start_date, end=end_date)
df_pdr
```

```python
# yahoo finance를 데이터 소스로 사용할 때는 종목코드 뒤에 코스피 종목인 경우 .KS, 코스닥 종목인 경우 .KQ를 붙여줘야 함
df_pdr = pdr.DataReader('005930.KS', 'yahoo', start=start_date, end=end_date)
df_pdr
```

## 2.4. yfinance

+ yahoo finance를 데이터 소스로 사용하는 패키지이기 때문에 종목코드 뒤에 코스피 종목인 경우 .KS, 코스닥 종목인 경우 .KQ를 붙여줘야 함
+ 가장 다양한 정보 확인 가능
+ 2000년 데이터부터 조회됨
+ end에 설정한 일자의 전일자까지 조회하기 때문에 end에 +1일을 해서 넣어줘야함

```python
ticker = yf.Ticker('005930.KS')
```

```python
ticker.history(
               interval='1d',
               start=start_date,
               end='2022-03-24',
               actions=True,
               auto_adjust=True)
```

```python
# 분단위 데이터 조회. 조회할 수 있는 기간에 제한이 있음
# 마지막행의 데이터는 당일 종가로 조회되는 듯?
df_minute = ticker.history(
               interval='1m',
               start='2022-03-20',
               end='2022-03-22',
               actions=True,
               auto_adjust=True)
df_minute
```

```python
# 재무정보도 조회가능
ticker.financials
```
