```python
from tqdm import tqdm
import pandas as pd
```

```python
no_data_list = []
for ticker in tqdm(krx_delisting.Symbol):
    # KRX delisting stock data 상장폐지된 종목 가격 데이터 (상장일~상장폐지일)
    try:
        fdr.DataReader(ticker, exchange='KRX-DELISTING').to_parquet(f'data_delisted/{ticker}.parquet')
    except:
        no_data_list += [ticker]
```
